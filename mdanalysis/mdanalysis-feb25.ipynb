```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the xvg files
data1 = np.loadtxt(r'C:\Users\paul\works\paulshamrat.github.io\mdanalysis\3unf100\em_potential.xvg')
data2 = np.loadtxt(r'C:\Users\paul\works\paulshamrat.github.io\mdanalysis\3unf-g210v-100\em_potential-1.xvg')

# Extract the columns from the data
x1 = data1[:,0]
y1 = data1[:,1]
x2 = data2[:,0]
y2 = data2[:,1]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the data
ax.plot(x1, y1, label='Data 1')
ax.plot(x2, y2, label='Data 2')

# Set the axis labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('em_potential')

# Add a legend
ax.legend()

# Save the plot as a PNG file
plt.savefig('em_potential.png')
```
